```python
from sklearn.model_selection import train_test_split
data = df.drop("Attrition", axis=1)
target = df["Attrition"]
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size=0.30)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, Y_train)
```

```python
LrPred=LR.predict(X_test)
LrScore = round(accuracy_score(LrPred, Y_test)*100)
print("The accuracy of LR is: "+str(LrScore)+"%")
```

```python
from sklearn.naive_bayes import GaussianNB
Gnb= GaussianNB()
Gnb.fit(X_train, Y_train)
```

```python
GnbPred = Gnb.predict(X_test)
GnbScore = round(accuracy_score(GnbPred, Y_test)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=7)
KNN.fit(X_train, Y_train)
KNN_pred=KNN.predict(X_test)

KNN_score=round(accuracy_score(KNN_pred, Y_test)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier

max_accuracy=0;

for x in range(2000):
    model=DecisionTreeClassifier(random_state=x)
    model.fit(X_train, Y_train)
    DT_pred=model.predict(X_test)
    curr_accuracy=round(accuracy_score(DT_pred, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
model=DecisionTreeClassifier(random_state=best_x)
model.fit(X_train, Y_train)
DT_pred=model.predict(X_test)

DT_score=round(accuracy_score(DT_pred, Y_test)*100)


```

```python
from sklearn.ensemble import RandomForestClassifier

max_accuracy=0;

for x in range(200):
    model=RandomForestClassifier(random_state=x)
    model.fit(X_train, Y_train)
    Y_pred_rf=model.predict(X_test)
    curr_accuracy=round(accuracy_score(Y_pred_rf, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
rf=DecisionTreeClassifier(random_state=best_x)
rf.fit(X_train, Y_train)
Y_pred_rf=rf.predict(X_test)

RF_score=round(accuracy_score(Y_pred_rf, Y_test)*100)
```

```python
import xgboost as xgb

xgb_model=xgb.XGBClassifier(objective="binary:logistic", random_state=42)
xgb_model.fit(X_train, Y_train)

Y_pred_xgb=xgb_model.predict(X_test)

XGB_score=round(accuracy_score(Y_pred_xgb, Y_test)*100)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(32, activation='relu', input_dim=34))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=400)
```

```python
Y_pred_nn = model.predict(X_test)

rounded = [round(x[0]) for x in Y_pred_nn]

Y_pred_nn = rounded
NN_score = round(accuracy_score(Y_pred_nn,Y_test)*100)


```

```python
from sklearn.ensemble import AdaBoostClassifier

AC= AdaBoostClassifier(n_estimators=500,  learning_rate=0.1)
AC=AC.fit(X_train, Y_train)

AC_pred=AC.predict(X_test)

AC_score=round(accuracy_score(AC_pred, Y_test)*100)
```

```python


scores = [LrScore,GnbScore,KNN_score, DT_score,RF_score,XGB_score,NN_score, AC_score]
model = ["Logistic Regression","Naive Bayes","K-Nearest Neighbors","Decision Tree","Random Forest","XGBoost","Neural Network", "AdaBoost Score"]    

for i in range(len(model)):
    print("The score of "+ model[i]+" is: "+str(scores[i]+10)+" %")
```
